```python
import time

n = 2000   # matrix size

start = time.perf_counter()

# Create matrices using list
l1 = [[i*n + j for j in range(n)] for i in range(n)]
l2 = [[i*n + j for j in range(n)] for i in range(n)]

# Element-wise addition
result = [[l1[i][j] + l2[i][j] for j in range(n)] for i in range(n)]

end = time.perf_counter()

print("List Time:", end - start)
```

```python
import numpy as np
import time

n = 2000   # matrix size

start = time.perf_counter()

m1 = np.arange(n*n).reshape(n, n)
m2 = np.arange(n*n).reshape(n, n)

result = m1 + m2

end = time.perf_counter()

print("NumPy Time:", end - start)
```

# Python Array Module

# The Python Array Module is a built-in module in Python used to create arrays that store elements of the same data type (unlike lists, which can store mixed types).


# It is useful when you want memory-efficient numeric storage.

```python
| Feature   | Array                      | List          |
| --------- | -------------------------- | ------------- |
| Data type | Same type only             | Mixed allowed |
| Memory    | Less memory                | More memory   |
| Speed     | Faster than list (numeric) | Slower        |
| Built-in  | Yes                        | Yes           |
```

```python
| Type Code | Data Type        |
| --------- | ---------------- |
| 'b'       | Signed char      |
| 'i'       | Signed integer   |
| 'I'       | Unsigned integer |
| 'f'       | Float            |
| 'd'       | Double float     |
```

# Array vs NumPy Array

```python
| Python Array                 | NumPy Array              |
| ---------------------------- | ------------------------ |
| Built-in module              | External library         |
| 1-D only                     | Multi-dimensional        |
| Limited operations           | Powerful math operations |
| Less popular in Data Science | Industry standard        |
```

```python
| Python Array                 | NumPy Array                      |
| ---------------------------- | -------------------------------- |
| Built-in module (`array`)    | External library (`numpy`)       |
| Only 1-Dimensional           | Multi-Dimensional (1D, 2D, 3D…)  |
| Limited operations           | Powerful mathematical operations |
| Less commonly used           | Industry standard                |
| Slower for large computation | Very fast (C optimized)          |
```
